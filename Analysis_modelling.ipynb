```python
X = data[["Years", "Job Rate"]]  # List of strings, not a tuple
Y = data["Annual Salary"]
```

```python
%pip install scikit-learn
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```python
len(Y_train)
```

```python
len(X_test)
```

```python
X
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,Y_train)
```

```python
lr.predict(X_test)
```

```python
prediction = lr.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mean_absolute_error(prediction,Y_test)
```

```python
# training extra models
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.metrics import root_mean_squared_error

import pandas as pd
import joblib
```

```python
# Random Forest Regressor
rf = RandomForestRegressor(random_state=42)
rf.fit(X_train, Y_train)

```

```python
# Gradient Boosting Regressor
gbr = GradientBoostingRegressor(random_state=42)
gbr.fit(X_train, Y_train)
```

```python
from sklearn.metrics import mean_absolute_error, r2_score, root_mean_squared_error

# Putting all models in a dictionary for easy comparison
models = {
    "Linear Regression": lr,
    "Random Forest": rf,
    "Gradient Boosting": gbr
}

# Initializing a list to store the results
results = []

# Looping through each model and calculate metrics
for name, model in models.items():
    y_pred = model.predict(X_test)
    mae = mean_absolute_error(Y_test, y_pred)
    rmse = root_mean_squared_error(Y_test, y_pred)  
    r2 = r2_score(Y_test, y_pred)
    
    results.append({
        "Model": name,
        "MAE": round(mae, 2),
        "RMSE": round(rmse, 2),
        "RÂ²": round(r2, 3)
    })

```

```python
# Convert the list of dictionaries into a pandas DataFrame
results_df = pd.DataFrame(results)

# Sort models by MAE (lower is better)
results_df = results_df.sort_values(by="MAE")

# Display results
print(results_df)
```

```python


# Save all models with filenames matching the Streamlit app
joblib.dump(lr, "linearmodel.pkl")
joblib.dump(rf, "randomforest_model.pkl")
joblib.dump(gbr, "gradientboosting_model.pkl")
```
